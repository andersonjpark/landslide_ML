```python
new_data = raw_data[["fatalities", "injuries","landslide_size", "trigger", "hazard_type", "country", "population", "latitude", "longitude"]]
```

```python
new_data = new_data.dropna(axis=0, how='any') # dropping all data points that has NaN in new_data
```

```python
new_data.plot(kind="scatter", x= "longitude", y = "latitude", alpha = 0.1,
             s = new_data["population"]/5000, label = "population", figsize = (20,12),
             c = "fatalities", cmap = plt.get_cmap("jet"), colorbar = True, sharex=False)
```
